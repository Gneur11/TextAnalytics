```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import re
```

```python
data = pd.read_json("orderedTwitterArchive.json")
data['created_at'] = pd.to_datetime(data.created_at, format="%Y-%m-%d %H:%M:%S")
data.sort_values(by=['created_at'], inplace=True, ascending=True)
data.reset_index(inplace=True,drop=True)
data
```

```python
#lowercase everything
data['modded_text'] = data['text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
```

```python
#remove hashtags, useless
for idx,row in data.iterrows():
    re.sub("^#\S+","",row["modded_text"])
data
```

```python
#check tags
ats = dict()
for idx,row in data.iterrows():
    l = re.findall("^@...[a-zA-Z0-9]*",row["modded_text"])
    for el in l:
        if el not in ats:
            ats[el] = 1
        else:
            ats[el] = ats[el]+1
    row["modded_text"] = row["modded_text"].replace("^@...[a-zA-Z0-9]*","")
print(len(ats))

#ats
```

```python
s = pd.Series(ats)
t = []
for el in s.index:
    if s[el] <= 2:
        t.append(el)
len(t)
```

```python
for el in t:
    s.drop(el,inplace=True)
#s
```

```python
#crea i dati
for tag in s.index:
    column = []
    for idx,row in data.iterrows():
        if tag in row["modded_text"].split():
            column.append(1)
        else:
            column.append(0)
    data[tag] = pd.Series(column)
```

```python
#remove punctuation
data['modded_text'] = data["modded_text"].str.replace("[^\w\s]",'')
data["modded_text"]
```

```python
#remove links
data["modded_text"] = data["modded_text"].str.replace("http\S+"," ")
```
